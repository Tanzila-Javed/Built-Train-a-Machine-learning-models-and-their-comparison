# **Machine Learning Model**

## **Load Data**

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Salary_dataset.csv')
df
```

##**Data Preparation**

### **Data Separation as X and Y**

```python
y=df['Salary']
y
```

```python
x=df.drop('Salary',axis=1)
x
```

### Data Spliting into training set(80%) and testing set(20%)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, )
```

```python
x_train
```

```python
x_test
```

## **Model Building**

### **Linear Regression**

**Training the model**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

### **Appling the model to make prediction**

```python
y_lr_train_pred= lr.predict(x_train)
y_lr_test_pred= lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

###**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse= mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2= r2_score(y_train, y_lr_train_pred)

lr_test_mse= mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2= r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results= pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns=['Method', 'Training MSE', 'Training r2' , 'Testing MSE', 'Testing r2']
lr_results
```

## **Random Forest**

##**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf= RandomForestRegressor()
rf.fit(x_train, y_train)
```

### **Applying the model to make prediction**

```python
y_rf_train_pred= rf.predict(x_train)
y_rf_test_pred= rf.predict(x_test)
```

## **Evaluate the Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse= mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2= r2_score(y_train, y_rf_train_pred)

rf_test_mse= mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2= r2_score(y_test, y_rf_test_pred)
```

```python
rf_train_mse
```

```python
rf_train_r2
```

```python
rf_test_mse
```

```python
rf_test_r2
```

```python
rf_results= pd.DataFrame(['RandomForestRegression', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns=['Method', 'Training MSE', 'Training r2' , 'Testing MSE', 'Testing r2']
rf_results
```

# **Support Vector Machine**

## **Training the model**

```python
from sklearn import svm

# Create an SVM model with a linear kernel
s = svm.SVC(kernel='linear')

# Train the model
s.fit(x_train , y_train)
```

### **Appling the model to make prediction**

```python
y_s_train_pred= s.predict(x_train)
y_s_test_pred= s.predict(x_test)
```

```python
print(y_s_train_pred)
```

```python
print(y_s_test_pred)
```

## **Evaluate the Model**

```python
from sklearn.metrics import mean_squared_error, r2_score
s_train_mse= mean_squared_error(y_train, y_s_train_pred)
s_train_r2= r2_score(y_train, y_s_train_pred)

s_test_mse= mean_squared_error(y_test, y_s_test_pred)
s_test_r2= r2_score(y_test, y_s_test_pred)
```

```python
s_train_mse
```

```python
s_train_r2
```

```python
s_test_mse
```

```python
s_test_r2
```

```python
s_results= pd.DataFrame(['SVM', s_train_mse, s_train_r2, s_test_mse , s_test_r2]).transpose()
s_results.columns=['Method', 'Training MSE', 'Training r2' , 'Testing MSE', 'Testing r2']
s_results
```

# **Concatenate the results of all Models**

```python
df_models=pd.concat([lr_results, rf_results,s_results], axis=0)
df_models
```
